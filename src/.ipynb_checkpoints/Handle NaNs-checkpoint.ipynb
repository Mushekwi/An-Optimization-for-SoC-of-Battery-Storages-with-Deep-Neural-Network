```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
pd.options.display.max_rows = 1000
```

```python
datapath = "../input/partially_cleaned"
    
load_filename = "Load_Amprion_2010-2017_cleaned.csv"
pvgeneration_filename = "PV_Generation_Amprion_2011-2017_cleaned.csv"

load = pd.read_csv(os.path.join(datapath, load_filename))
pvgeneration = pd.read_csv(os.path.join(datapath, pvgeneration_filename))
```

# Load data

```python
def get_nans_above_threshold(df,colname, na_threshold):
    count = 0
    na_idxs_above_threshold = np.array([])
    current_nas = []
    for (idx,row) in df.iterrows():
        # count na's
        if np.isnan(row[colname]):
            count += 1
            current_nas.append(idx)
        else:
            if len(current_nas) >= na_threshold: # if above certain threshold
                na_idxs_above_threshold = np.append(na_idxs_above_threshold,np.array(current_nas))
            count = 0
            current_nas = []

    return np.array(na_idxs_above_threshold).flatten()
```

```python
# load has some negative values.
# change them to 0.
load.loc[load['actual'] < 0 , 'actual'] = 0.0

# set all 0s in load.actual to load.forecast
load.loc[load['actual'] == 0, 'actual'] = load.loc[load['actual'] == 0, 'forecast']

# iloc 78151 has totally wrong data. set it to np.nan
load.loc[78151,'actual'] = np.nan


# sometimes whole day is missing.
# we can copy data from one week later.


na_above_threshold = get_nans_above_threshold(load, 'actual', 10)    
na_plus_1week_idxs = na_above_threshold + 7*24*4 # 7 days, 24 hours, 4 quarter
load.loc[na_above_threshold, 'actual'] = load.loc[na_plus_1week_idxs, 'actual'].values
load.loc[na_above_threshold, 'forecast'] = load.loc[na_plus_1week_idxs, 'forecast'].values


# interpolate np.nan values
# but we cannot interpolate whole day. so set limit to 10
load = load.interpolate()
```

```python
load.iloc[:].plot(figsize=(20,10),alpha=0.5)
    
```

```python
assert load.isnull().sum().sum() == 0
```

# PV Generation

```python
pvgeneration.loc[pvgeneration.forecast.isnull()]
```

```python
pvgeneration.loc[24800-300:24800+300,['actual','forecast']].plot(figsize=(20,10),alpha=0.5)
```

```python
# change dtype to datetime
pvgeneration[['from','to']] = pvgeneration[['from','to']].apply(lambda x:pd.to_datetime(x, format='%Y-%m-%d %H:%M'))
```

```python
# set 0 to hour -> 21:00-05:00
_21_00 = pd.to_datetime('21:00:00', format='%H:%M:%S').time()
_05_00 = pd.to_datetime('05:00:00', format='%H:%M:%S').time()
pvgeneration.loc[pvgeneration['from'].apply(lambda x: x.time() >  _21_00), 'actual'] = 0.0
pvgeneration.loc[pvgeneration['from'].apply(lambda x: x.time() >  _21_00), 'forecast'] = 0.0
pvgeneration.loc[pvgeneration['from'].apply(lambda x: x.time() <  _05_00), 'actual'] = 0.0
pvgeneration.loc[pvgeneration['from'].apply(lambda x: x.time() <  _05_00), 'forecast'] = 0.0
```

```python
# rather than 7 days later, we copied the day after values.
# because pvgeneration is more correlated in consequent days
na_above_threshold_actual = get_nans_above_threshold(pvgeneration, 'actual', 10)    
                                                                        # 1 day, 24 hours, 4 quarter
pvgeneration.loc[na_above_threshold_actual, 'actual'] = pvgeneration.loc[na_above_threshold_actual + 1*24*4, 'actual'].values

na_above_threshold_forecast = get_nans_above_threshold(pvgeneration, 'forecast', 10) 
pvgeneration.loc[na_above_threshold_forecast, 'forecast'] = pvgeneration.loc[na_above_threshold_forecast + 1*24*4, 'forecast'].values
```

```python
# set np.nan values of actual to forecast if forecast is 0.0
pvgeneration.loc[(pvgeneration['actual'].isnull()) & (pvgeneration['forecast'] == 0.0), 'actual'] = 0
# set np.nan values of forecast to actual if actual is 0.0
pvgeneration.loc[(pvgeneration['forecast'].isnull()) & (pvgeneration['actual'] == 0.0), 'forecast'] = 0
```

```python
# set np.nan values of actual to forecast
pvgeneration.loc[pvgeneration['actual'].isnull(), 'actual'] = pvgeneration.loc[pvgeneration['actual'].isnull(), 'forecast']
# set np.nan values of forecast to actual
pvgeneration.loc[pvgeneration['forecast'].isnull(), 'forecast'] = pvgeneration.loc[pvgeneration['forecast'].isnull(), 'actual']
```

```python
# interpolate other np.nans
pvgeneration = pvgeneration.interpolate()
```

```python
assert pvgeneration.isnull().sum().sum() == 0
```

# Save Load and PVGeneration

```python
load.to_csv("../input/load.csv", index=None)
pvgeneration.to_csv("../input/pvgeneration.csv", index=None)
```
