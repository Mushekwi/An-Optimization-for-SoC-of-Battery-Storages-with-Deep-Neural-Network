```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
pd.options.display.max_rows = 1000
```

```python
datapath = "../input"
    
load_filename = "Load_Amprion_2010-2017_cleaned.csv"
pvgeneration_filename = "PV_Generation_Amprion_2011-2017_cleaned.csv"

load = pd.read_csv(os.path.join(datapath, load_filename))
pvgeneration = pd.read_csv(os.path.join(datapath, pvgeneration_filename))
```

# Load data

```python
# load has some negative values.
# change them to 0.
load.loc[load['actual'] < 0 , 'actual'] = 0.0

# set all 0s in load.actual to load.forecast
load.loc[load['actual'] == 0, 'actual'] = load.loc[load['actual'] == 0, 'forecast']

# iloc 78151 has totally wrong data. set it to np.nan
load.loc[78151,'actual'] = np.nan


# sometimes whole day is missing.
# we can copy data from one week later.
def get_nans_above_threshold(df,colname, threshold):
    na_threshold = 10
    count = 0
    na_idxs_above_threshold = []
    current_nas = []
    for (idx,row) in load.iterrows():
        # count na's
        if np.isnan(row['actual']):
            count += 1
            current_nas.append(idx)
        else:
            if len(current_nas) >= na_threshold: # if above certain threshold
                na_idxs_above_threshold.append(current_nas)
            count = 0
            current_nas = []

    return np.array(na_idxs_above_threshold).flatten()

na_above_threshold = get_nans_above_threshold(load, 'actual', 10)    
na_plus_1week_idxs = na_above_threshold + 7*24*4 # 7 days, 24 hours, 4 quarter
load.loc[na_above_threshold, 'actual'] = load.loc[na_plus_1week_idxs, 'actual'].values
load.loc[na_above_threshold, 'forecast'] = load.loc[na_plus_1week_idxs, 'forecast'].values


# interpolate np.nan values
# but we cannot interpolate whole day. so set limit to 10
load = load.interpolate()
```

```python
load.iloc[:].plot(figsize=(20,10),alpha=0.5)
    
```

```python
# rather than 7 days later, we copied the day after values.
# because pvgeneration is more correlated in consequent days
na_above_threshold = get_nans_above_threshold(pvgeneration, 'actual', 10)    
na_plus_1week_idxs = na_above_threshold + 1*24*4 # 1 day, 24 hours, 4 quarter
pvgeneration.loc[na_above_threshold, 'actual'] = pvgeneration.loc[na_plus_1week_idxs, 'actual'].values
pvgeneration.loc[na_above_threshold, 'forecast'] = pvgeneration.loc[na_plus_1week_idxs, 'forecast'].values
```
